```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
```

```python
fft = pd.read_csv('data/grafana_data_export oct_fftcb.csv', sep=';', engine='python', header=1)
fft['Time'] = pd.to_datetime(fft['Time'])

for band in fft.Series.unique():
    figure(figsize=(20, 6), dpi=80)
    plt.plot(fft[fft['Series'] == band]['Value'])
    plt.title(band)
    plt.ylabel('FFT')
    plt.show()

```
